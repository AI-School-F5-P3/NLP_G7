```python
import matplotlib.pyplot as plt

# Datos mockeados
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Crear la gráfica
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Título y etiquetas
plt.title('Mocked Data Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Mostrar la gráfica
plt.grid(True)
plt.show()
```
